```python
import string
#Using string library define a function that replaces any punctuations present in a string with a blankspace 
def replace_punctuation(input_string):
    # Make a translation table
    translator = str.maketrans(string.punctuation, " " * len(string.punctuation))

    # Use this table to replace all punctuation with spaces
    no_punct = input_string.translate(translator)
    return no_punct
#Prints Semester 'Number' or prints 'No semester info
def searchnprint(document):
    for notice in document:
        notice = notice.replace(" ","")
        present = 0 
        for idx in range(0,8):
            if any(x in notice for x in bow[idx]):
                print('{}'.format(idx+1))
                present = 1      
        for idx in range(0,7):
            if any(x in notice for x in bow2[idx]):
                print('{}'.format(idx+1)) 
                present = 1
        if not present:
            print('No semester info found')   
#function to lower case and replace 'semester' with 'sem'
def function(list):
    for idx, each in enumerate(list):
        list[idx] = (replace_punctuation(each.text)).lower().replace('semester','sem')
# Prints list and its index. Each element is in new line
def disp(list):
    for idx, item in enumerate(list):
        print('({}) {}'.format(idx,item)) 
```

```python
#loop to find idx of all sem words in the list
#Note maximum length of list is 3. Not more than 3sems will be found
def find_idx(text):
    pos=0
    i = 0
    idx = [-2,-2,-2]
    while (pos < len(text)) & (i < 4):
        text = text[pos:]
        if text.find('sem') == -1:
            break
        idx[i] = text.find('sem')
        pos = idx[i] + len('sem')
        i = i+1
    #print(idx)    
    return idx    
#remove words: 'and', 'for', 'by'
def remove_stopwords(text):
    stopwords = ['and', 'for', 'by','of']
    querywords = text.split()
    #print(querywords)
    result_words = [word for word in querywords if word not in stopwords]
    result = ' '.join(result_words)
    #print(result)
    return(result)

def sem_nos(Alist):
    for each in list:
        sem =''
        each = remove_stopwords(each)
        each = each.replace(" ","") #takes some time to run
        idx = find_idx(each)
        for i in [0,1,2]:
            if (idx[i] !=-2):
                A = each[idx[i]-20:idx[i]] + each[idx[i]:idx[i]+23]
                for char in A:
                    if char.isnumeric():
                        sem = sem + char
        if (len(sem) ==0):
            print('No Semester info found')
        else: 
            print(sem)  

def roman(Alist):
    for text in Alist:
        check_list = ['i','ii','iii','iv','v','vi','vii','viii'] 
        text_list = text.split()
        # result = [roman for roman in set(check_list).intersection(text_list)]
        result = [{roman : text_list.index(roman)} for roman in check_list if roman in text_list]
        for dict in result:
            for key in dict:
                search_idx = dict[key]
                if 'sem' not in text_list[search_idx-3:search_idx+4]:
                    result.remove(dict) 
        print(result)           
```
